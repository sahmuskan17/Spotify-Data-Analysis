Data Analysis

###**Top 5 most popular artist**

```python
top_five_artist = df.groupby("artist").count().sort_values(by ="song_title", ascending = False)["song_title"][:5]
top_five_artist
```

```python
top_five_artist.plot(kind='bar', figsize=(10, 4))
plt.title("Top 5 Most Popular Artist")
plt.show()
```

###**Top 5 loudest track**

```python
top_five_tracks = df[["song_title","loudness"]].sort_values(by ="loudness", ascending = True)[:5]
top_five_tracks
```

```python
plt.figure(figsize = (10,5))
colors = ["pink","red","blue","green","yellow"]
sns.barplot(x="loudness", y="song_title", data=top_five_tracks, hue="song_title", palette=colors, dodge=False, legend=False)
plt.title("Top 5 Loudest Track")
plt.show()
```

```python
top_danceability_song = df[["artist","danceability","song_title"]].sort_values(by = "danceability", ascending = False)[:5]
top_danceability_song
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x ="danceability", y="artist", data=top_danceability_song, hue = "song_title")
plt.title("Top 5 Danceability Song")
plt.show()
```

###**Top 5 instrumental tracks**

```python
top_ten_instrumental_tracks = df[["instrumentalness", "song_title", "artist"]].sort_values(by = "instrumentalness", ascending = False)[:5]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize = (10,5))
plt.pie(x = "instrumentalness", data = top_ten_instrumental_tracks, labels ="song_title", autopct = "%1.2f%%")
plt.title("Top 5 Instrumental Tracks")
plt.show()
```

```python
interest_feature_cols = ["danceability", "energy", "loudness", "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo","duration_ms" ]
```

```python
for feature_col in interest_feature_cols:
  pos_data = df[df["target"]== 1][feature_col]
  neg_data = df[df["target"]== 1][feature_col]


  plt.figure(figsize= (10,5))
  sns.displot(pos_data, bins = 30, label = "positive", color = "green")
  sns.displot(neg_data, bins = 30, label = "negative", color = "red")
  plt.title(f"Distribution of {feature_col}")

  plt.legend(loc = "upper right")
  plt.show()


```
